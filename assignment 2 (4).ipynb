```python
  #Answer : 2
    
import pandas as pd

def reindex_with_increment(df):
    # Create a new index starting from 1 and incrementing by 2
    new_index = range(1, len(df) * 2, 2)
    
    # Assign the new index to the DataFrame
    df.index = new_index
    
    return df

# Sample DataFrame 'df' with columns 'A', 'B', and 'C'
data = {'A': [10, 20, 30],
        'B': [40, 50, 60],
        'C': [70, 80, 90]}

df = pd.DataFrame(data)

# Call the function to re-index the DataFrame
new_df = reindex_with_increment(df)

# Display the DataFrame with the new index
print(new_df)
```

```python
    #Answer : 3
    
import pandas as pd

def calculate_sum_of_first_three(df):
    # Check if the 'Values' column exists in the DataFrame
    if 'Values' in df.columns:
        # Calculate the sum of the first three values in the 'Values' column
        total_sum = df['Values'].head(3).sum()
        
        # Print the sum to the console
        print("Sum of the first three values:", total_sum)
    else:
        print("DataFrame does not contain a 'Values' column.")

# Sample DataFrame 'df' with a 'Values' column
data = {'Values': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Call the function to calculate and print the sum of the first three values
calculate_sum_of_first_three(df)
```

```python
    #Answer : 4
    
import pandas as pd

def count_words_in_text(df):
    # Function to count words in a text
    def count_words(text):
        words = text.split()
        return len(words)

    # Apply the count_words function to each row in the 'Text' column
    df['Word_Count'] = df['Text'].apply(count_words)

    return df

# Sample DataFrame 'df' with a 'Text' column
data = {'Text': ['This is a sample sentence.',
                 'Count the words in this text.',
                 'Pandas is a powerful library.']}

df = pd.DataFrame(data)

# Call the function to add a 'Word_Count' column
df = count_words_in_text(df)

# Display the DataFrame with the new 'Word_Count' column
print(df)
    
```

```python
    #Answer : 7
    
import pandas as pd

def extract_username_from_email(df):
    # Function to extract the username from an email address
    def extract_username(email):
        # Split the email address using '@' as a separator and get the first part (username)
        username = email.split('@')[0]
        return username

    # Apply the extract_username function to each row in the 'Email' column
    df['Username'] = df['Email'].apply(extract_username)

    return df

# Sample DataFrame 'df' with an 'Email' column
data = {'Email': ['john.doe@example.com', 'jane.smith@example.com', 'bob@example.com']}

df = pd.DataFrame(data)

# Call the function to add a 'Username' column
df = extract_username_from_email(df)

# Display the DataFrame with the new 'Username' column
print(df)
    
```

```python
    #Answer : 8
    
import pandas as pd

def select_rows(df):
    # Define the conditions for selection
    condition = (df['A'] > 5) & (df['B'] < 10)
    
    # Use boolean indexing to select rows that meet the conditions
    selected_rows = df[condition]
    
    return selected_rows

# Sample DataFrame 'df' with columns 'A', 'B', and 'C'
data = {'A': [3, 8, 6, 2, 9],
        'B': [5, 2, 9, 3, 1],
        'C': [1, 7, 4, 5, 2]}

df = pd.DataFrame(data)

# Call the function to select rows based on the conditions
selected_df = select_rows(df)

# Display the selected DataFrame
print(selected_df)
    
```

```python
    #Answer : 9
    
import pandas as pd

def calculate_statistics(df):
    # Calculate the mean, median, and standard deviation of the 'Values' column
    mean_value = df['Values'].mean()
    median_value = df['Values'].median()
    std_deviation = df['Values'].std()
    
    return mean_value, median_value, std_deviation

# Sample DataFrame 'df' with a 'Values' column
data = {'Values': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Call the function to calculate statistics
mean, median, std = calculate_statistics(df)

# Print the calculated statistics
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)
    
```

```python
    #Answer : 10
    
import pandas as pd

def calculate_moving_average(df):
    # Sort the DataFrame by the 'Date' column (if not already sorted)
    df = df.sort_values(by='Date')
    
    # Calculate the moving average with a window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    
    return df

# Sample DataFrame 'df' with 'Date' and 'Sales' columns
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', '2023-01-06', '2023-01-07'],
        'Sales': [10, 15, 20, 25, 30, 35, 40]}

df = pd.DataFrame(data)

# Convert the 'Date' column to datetime if it's not already
df['Date'] = pd.to_datetime(df['Date'])

# Call the function to calculate the moving average
df = calculate_moving_average(df)

# Display the DataFrame with the new 'MovingAverage' column
print(df)
    
```

```python
    #Answer : 11
    
import pandas as pd

def add_weekday_column(df):
    # Convert the 'Date' column to a datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Create the 'Weekday' column containing weekday names
    df['Weekday'] = df['Date'].dt.strftime('%A')
    
    return df

# Sample DataFrame 'df' with a 'Date' column
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}

df = pd.DataFrame(data)

# Call the function to add the 'Weekday' column
df = add_weekday_column(df)

# Display the modified DataFrame
print(df)
    
```

```python
    #Answer : 12
    
import pandas as pd

def select_rows_in_date_range(df, start_date, end_date):
    # Convert the 'Date' column to datetime if it's not already
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Define the conditions for selecting rows within the date range
    condition = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    
    # Use boolean indexing to select rows that meet the conditions
    selected_rows = df[condition]
    
    return selected_rows

# Sample DataFrame 'df' with a 'Date' column
data = {'Date': ['2023-01-01', '2023-01-15', '2023-01-31', '2023-02-10', '2023-02-20']}

df = pd.DataFrame(data)

# Call the function to select rows within the date range '2023-01-01' to '2023-01-31'
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2023-01-31')
selected_df = select_rows_in_date_range(df, start_date, end_date)

# Display the selected DataFrame
print(selected_df)
  
```
